```python
from tools.functions import *
import pickle
```

```python
path_to_data = "final_data/"

# Add models to compare
condition_0 = "base"
condition_1 = "sparse_high_start"
condition_2 = "sparse_age_01_born"
condition_3 = "sparse_age_01_born_threshold_turnover"


conditions = [condition_0,
              condition_1, 
              condition_2, 
              condition_3
              ]
```

```python
# plot proactiv interference

colors = [[0.216, 0.639, 0.82], [0.941, 0.62 , 0.137], [0.875, 0.22 , 0.09], [0.322, 0.714, 0.345], [0.788, 0.373, 0.773]]

space = -0.15
j = 0

for condition in conditions:
    with open(path_to_data + condition + "_aggregated_outputs.pkl", "rb") as f:
        data = pickle.load(f)
    
    means_between = np.mean(data["pm_table_between"], axis=1)
    stds_between = np.std(data["pm_table_between"], axis=1)

    plt.errorbar(np.arange(10)+space, means_between, stds_between, linestyle='None', marker='o', color=colors[j])
    plt.xlabel("Group")
    plt.ylabel("Percent match")
    plt.title("Percent match for each group with binarized data (Proactive interference)")
    plt.ylim(0.75, 1)
    space += 0.15

    j += 1
plt.legend(conditions, loc="lower left")
plt.show()

space = -0.15
j = 0

for condition in conditions:
    with open(path_to_data + condition + "_aggregated_outputs.pkl", "rb") as f:
        data = pickle.load(f)
    
    means_within = np.mean(data["pm_table_within"], axis=1)
    stds_within = np.std(data["pm_table_within"], axis=1)

    plt.errorbar(np.arange(10)+space, means_within, stds_within, linestyle='None', marker='o', color=colors[j])
    plt.xlabel("Group")
    plt.ylabel("Match")
    plt.title("Match for each group (Retroactive interference)")
    plt.ylim(0.75, 1)
    space += 0.15
    j += 1
plt.legend(conditions, loc="lower left")  
plt.show()
    
```
